## 5. Store this dataset into mongodb

```python
import pymongo
import json
from pymongo import InsertOne

url = "mongodb://localhost:27017/"

db_name = "attribute_dataset"

client = pymongo.MongoClient(url)

dataBase = client[db_name]

# creating collection inside the database (collection is nothing but the table)
collection = dataBase["attribute_dataset"]

requesting = []

# storing table attribute dataset
with open(r"/Users/rajaiswal/Desktop/learning/Fullstack_Datascience_Ineuron/Tasks/attribute_dataset.json") as f:
    for jsonObj in f:
        myDict = json.loads(jsonObj)
        requesting.append(InsertOne(myDict))

result = collection.bulk_write(requesting)
print("Sucessful")
client.close()
```

## 6. Perform left join operation with attrubute dataset and dress dataset on column Dress_ID

```python
import mysql.connector as conn

mydb = conn.connect(host="localhost", user="root", password="rootroot")  # to establish the databases
cursor = mydb.cursor()  # to create a cursor in database

l = []
cursor.execute('use attribute_dataset')

# query to perform left join
q1 = 'select dress_details.Dress_ID, dress_details.style, dress_details.price from dress_details left join dress_sales on  dress_details.Dress_ID = dress_sales.Dress_ID order by dress_details.Dress_ID;'
cursor.execute(q1)

for i in cursor.fetchall():
    l.append(i)

print(l)
```

## 7. Write a sql query to find out how many unique dress that we have based on dress id
