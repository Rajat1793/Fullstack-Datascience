### READING AGENT PERFORMANCE DATA

```python
performance = pd.read_csv("/Users/rajaiswal/Desktop/learning/Fullstack_Datascience_Ineuron/DataFiles/AgentPerformance.csv")
```

```python
performance.head()
```

```python
performance.dtypes
```

```python
# convert date into pandas date to extract week
performance['Date'] = pd.to_datetime(performance['Date'])
performance['Week'] = pd.to_datetime(performance['Date']).dt.week
```

```python
performance.dtypes
```

```python
performance["day"] = performance['Date'].apply (lambda r :r.day)
performance["month"] = performance['Date'].apply (lambda r :r.month)
performance["year"] = performance['Date'].apply (lambda r :r.year)
performance["week"] = performance['Date'].apply (lambda r :r.week)
```

```python
performance.head()
```
