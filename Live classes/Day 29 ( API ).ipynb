```python
from flask import Flask, request, jsonify
import mysql.connector as conn

mydb = conn.connect(host="localhost", user="root", password="rootroot")  # to establish the databases
cursor = mydb.cursor()

# app is object for the class flask
app = Flask(__name__)


# function decorator
# @app.route('/abc', methods=['GET', 'POST'])
# def test_api():
#     if request.method == 'POST':
#         a = request.json['num1']
#         b = request.json['num2']
#         result = a + b
#         return jsonify((str(result)))
#
#
# @app.route('/abcd', methods=['GET', 'POST'])
# def test_api2():
#     if request.method == 'POST':
#         a = request.json['num1']
#         b = request.json['num2']
#         result = a * b
#         return jsonify((str(result)))
#

@app.route('/showdata', methods=['GET'])
def showData():
    if request.method == 'GET':
        cursor.execute("Select * from student.details")
        return cursor.fetchall()


@app.route('/insertrecord', methods=['POST'])
def insertData():
    if request.method == 'POST':
        id = request.json["id"]
        name = request.json["name"]
        email = request.json["email"]
        clas = request.json["class"]
        phone = request.json["phone"]
        cursor.execute("use student")
        q = f'insert into student.details values({id},"{name}","{email}","{clas}", {phone});'
        cursor.execute(q)
        mydb.commit()
        return 'Data inserted'


@app.route('/deleterecord', methods=['DELETE'])
def deleteRecord():
    if request.method == 'DELETE':
        id = request.json["id"]
        q = f'delete from student.details where stud_id = {id}'
        cursor.execute(q)
        mydb.commit()
        return 'Record deleted'


@app.route('/updaterecord', methods=['POST'])
def updateRecord():
    if request.method == 'POST':
        id = request.json["id"]
        name = request.json["name"]
        cursor.execute("update student.details set name = %s where stud_id = %s", (name,id,))
        mydb.commit()
        return "cursor.fetchall()"


if __name__ == '__main__':
    app.run()
```
