```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df6 = pd.DataFrame(data6)
```

```python
df7 = pd.DataFrame(data7)
```

```python
# of we have to merge two dataframes with different row name we can do that using left on and right on
# by default value to how is inner

pd.merge(df6, df7, left_on='emp_id1', right_on='emp_id2')
```

```python
pd.merge(df6, df7, left_on='emp_id1', right_on='emp_id2', how = 'outer')
```

```python
def profit_flag(a):
    if a > 0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(24)
```

```python
df12['flag_profit'] = df12['profit'].apply(profit_flag)
```

```python
df12.head()
```

```python
df12['customer_name_len'] = df12['customer_name'].apply(len)
```

```python
def quantity_flag(a):
    if a < 10:
        return 'low'
    elif a >10 and a < 20:
        return 'medium'
    else:
        return 'high'
```

```python
df12['flag_quantity'] = df12['quantity'].apply(quantity_flag)
```

```python
df12['square_quantity'] = df12['quantity'].apply(lambda a : a**2)
```
